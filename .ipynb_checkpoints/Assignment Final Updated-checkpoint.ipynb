### Numerical conversion
* We convert all strings values that have **Yes** , **No** and **'others'** to numerical value. Than, we show data state after conversion. To check what is the unique string contains inside columns value, we call **unique** function. Finally, we show **correlation** heatmap between numerical values in dataset using ***Seaborn***

```python
data_corr = data.loc[:, "gender":"Churn"]

#Define a generic function using Pandas replace function
def coding(col, codeDict):
  colCoded = pd.Series(col, copy=True)
  for key, value in codeDict.items():
    colCoded.replace(key, value, inplace=True)
  return colCoded
 
#Coding to numerical value
data_corr["PhoneService"] = coding(data_corr["PhoneService"], {'No':0,'Yes':1})
data_corr["Partner"] = coding(data_corr["Partner"], {'No':0,'Yes':1})
data_corr["Dependents"] = coding(data_corr["Dependents"], {'No':0,'Yes':1})
data_corr["PaperlessBilling"] = coding(data_corr["PaperlessBilling"], {'No':0,'Yes':1})
#######
data_corr["gender"] = coding(data_corr["gender"], {'Male':0,'Female':1})
#######
data_corr["MultipleLines"] = coding(data_corr["MultipleLines"], {'No':0,'Yes':1, 'No phone service':2})
#######
data_corr["OnlineSecurity"] = coding(data_corr["OnlineSecurity"], {'No':0,'Yes':1, 'No internet service':2})
data_corr["OnlineBackup"] = coding(data_corr["OnlineBackup"], {'No':0,'Yes':1, 'No internet service':2})
data_corr["DeviceProtection"] = coding(data_corr["DeviceProtection"], {'No':0,'Yes':1, 'No internet service':2})
data_corr["TechSupport"] = coding(data_corr["TechSupport"], {'No':0,'Yes':1, 'No internet service':2})
data_corr["StreamingTV"] = coding(data_corr["StreamingTV"], {'No':0,'Yes':1, 'No internet service':2})
data_corr["StreamingMovies"] = coding(data_corr["StreamingMovies"], {'No':0,'Yes':1, 'No internet service':2})
#######
data_corr["InternetService"] = coding(data_corr["InternetService"], {'No':0,'DSL':1, 'Fiber optic':2})
#######
data_corr["Contract"] = coding(data_corr["Contract"], {'Month-to-month':0, 'One year':1, 'Two year':2})
#######
data_corr["PaymentMethod"] = coding(data_corr["PaymentMethod"], {'ELECTRONIC CHECK':0, 'MAILED CHECK':1, 'BANK TRANSFER (AUTOMATIC)':2,'CREDIT CARD (AUTOMATIC)':3})
#######
data_corr["Churn"] = coding(data_corr["Churn"], {'No':0,'Yes':1})
print data_corr.info()
```

### Display correlated attribute pair
We would like to see the attribute pair correlation value; how each attribute related to others.


```python
pearson = data_corr.corr(method='pearson')

attrs = pearson.iloc[:-1,:-1] 
threshold = 0.5
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]) \
    .unstack().dropna().to_dict()
    
unique_important_corrs = pd.DataFrame(
    list(set([(tuple(sorted(key)), important_corrs[key]) \
    for key in important_corrs])), columns=['attribute pair', 'correlation'])

# sorted by absolute value
unique_important_corrs = unique_important_corrs.ix[
    abs(unique_important_corrs['correlation']).argsort()[::-1]]
unique_important_corrs
```

### Train and Test set using scikit-learn libraries
Cross validation attempts to avoid overfitting (training on and predicting the same datapoint) while still producing a prediction for each observation dataset. This is accomplished by systematically hiding different subsets of the data while training a set of models. After training, each model predicts on the subset that had been hidden to it, emulating multiple train-test splits. When done correctly, every observation will have a 'fair' corresponding prediction.

```python
from sklearn.cross_validation import KFold

def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True)
    y_pred = y.copy()

    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

### Prediction algorithm

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.neighbors import KNeighborsClassifier as KNN

def accuracy(y_true,y_pred):
    # NumPy interprets True and False as 1. and 0.
    return np.mean(y_true == y_pred)

print "Support vector machines:"
print "%.3f" % accuracy(y, run_cv(X,y,SVC))
print "Random forest:"
print "%.3f" % accuracy(y, run_cv(X,y,RF))
print "K-nearest-neighbors:"
print "%.3f" % accuracy(y, run_cv(X,y,KNN))
```

From the accuracy we can see that SVM produce higher value of 0.804 compared to others.
### Confusion Matrix
We use confusion matrix to see the classes of True Positive / Negative and False Positive / Negative by comparing all 3 algorithms.

```python
#Declare neccessary function for confusion matrix
from sklearn.metrics import confusion_matrix

y = np.array(y)
class_names = np.unique(y)

import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y,run_cv(X,y,SVC))
cnf_matrix2 = confusion_matrix(y,run_cv(X,y,RF))
cnf_matrix3 = confusion_matrix(y,run_cv(X,y,KNN))
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Support Vector Machine')
plt.show()
plot_confusion_matrix(cnf_matrix2, classes=class_names,
                      title='Random Forest')
plt.show()
plot_confusion_matrix(cnf_matrix3, classes=class_names,
                      title='KNN')
plt.show()
```
